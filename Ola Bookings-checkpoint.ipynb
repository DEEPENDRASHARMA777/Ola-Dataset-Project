# âœ… ðŸ“Œ Step 1: Import necessary libraries

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# âœ… ðŸ“Œ Step 2: Load the dataset

```python
# Step 2: Load the dataset
file_path = 'Ola_ride_data.csv.xlsx'  # Ensure the file is in the same directory or provide full path
df = pd.read_excel(file_path, sheet_name='July')

# Step 3: Display first few rows
df.head()
```

# âœ… ðŸ“Œ Step 4: Basic Info & Null Values

```python
# Step 4: Basic Info & Null Values
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

# âœ… ðŸ“Œ Step 5: Summary statistics

```python
# Step 5: Summary statistics
df.describe(include='all')
```

# âœ… ðŸ“Œ Step 6: Value counts of Booking_Status

```python
# Step 6: Value counts of Booking_Status
print("\nBooking Status Distribution:\n", df['Booking_Status'].value_counts())
```

# âœ… ðŸ“Œ Step 7: Clean Data - Drop rows with all NaN values

```python
# Step 7: Clean Data - Drop rows with all NaN values
df.dropna(how='all', inplace=True)
```

# âœ… ðŸ“Œ Step 8: Handle missing Payment_Methods (optional: fill with 'Unknown')

```python
# Step 8: Handle missing Payment_Methods (optional: fill with 'Unknown')
df['Payment_Method'] = df['Payment_Method'].fillna('Unknown')
```

# âœ… ðŸ“Œ Step 9: Convert 'Date' to datetime (if needed)

```python
# Step 9: Convert 'Date' to datetime (if needed)
df['Date'] = pd.to_datetime(df['Date'])
```

# âœ… ðŸ“Œ Step 10: Plot booking status count

```python
# Step 10: Plot booking status count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Booking_Status', order=df['Booking_Status'].value_counts().index)
plt.title('Booking Status Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# âœ… ðŸ“Œ Step 11: Average ratings by vehicle type

```python
# Step 11: Average ratings by vehicle type
avg_ratings = df.groupby('Vehicle_Type')[['Driver_Ratings', 'Customer_Rating']].mean()
print("\nAverage Ratings by Vehicle Type:\n", avg_ratings)
```

# âœ… ðŸ“Œ Step 12: Total Booking Value by Payment Method

```python
# Step 12: Total Booking Value by Payment Method
payment_value = df.groupby('Payment_Method')['Booking_Value'].sum().sort_values(ascending=False)
print("\nTotal Booking Value by Payment Method:\n", payment_value)
```

# âœ… ðŸ“Œ Step 13: Plot Booking Value by Payment Method

```python
# Step 13: Plot Booking Value by Payment Method
payment_value.plot(kind='bar', color='skyblue', title='Total Booking Value by Payment Method')
plt.ylabel('Booking Value')
plt.tight_layout()
plt.show()
```

# âœ… ðŸ“Œ Step 14: Analyze Cancellation Reasons by Customer vs Driver

```python
# Count cancellation reasons by customer
cancel_by_customer = df['Canceled_Rides_by_Customer'].value_counts().dropna()
print("\nTop Reasons for Customer Cancellations:\n", cancel_by_customer)

# Count cancellation reasons by driver
cancel_by_driver = df['Canceled_Rides_by_Driver'].value_counts().dropna()
print("\nTop Reasons for Driver Cancellations:\n", cancel_by_driver)

# Plot both cancellation reason counts side-by-side
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
cancel_by_customer.plot(kind='barh', color='salmon')
plt.title('Customer Cancellation Reasons')
plt.xlabel('Count')

plt.subplot(1, 2, 2)
cancel_by_driver.plot(kind='barh', color='lightblue')
plt.title('Driver Cancellation Reasons')
plt.xlabel('Count')

plt.tight_layout()
plt.show()
```

# âœ… Step 15: Correlation Heatmap (Numerical Data)

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Booking_Value', 'Ride_Distance', 'V_TAT', 'C_TAT', 'Driver_Ratings', 'Customer_Rating']].corr(), 
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

# âœ… Step 16: Ride Distance vs Booking Value (Scatter Plot)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Ride_Distance', y='Booking_Value', hue='Vehicle_Type', palette='Set2')
plt.title('Ride Distance vs Booking Value')
plt.xlabel('Ride Distance (km)')
plt.ylabel('Booking Value (â‚¹)')
plt.tight_layout()
plt.show()
```

# âœ… Step 17: Average Booking Value by Vehicle Type

```python
vehicle_value = df.groupby('Vehicle_Type')['Booking_Value'].mean().sort_values(ascending=False)
print("\nAverage Booking Value by Vehicle Type:\n", vehicle_value)

vehicle_value.plot(kind='bar', color='orange', title='Avg Booking Value by Vehicle Type')
plt.ylabel('Average â‚¹')
plt.tight_layout()
plt.show()
```

# âœ… Step 18: Daily Booking Trend

```python
# Count bookings per day
daily_bookings = df['Date'].dt.date.value_counts().sort_index()

plt.figure(figsize=(10, 4))
daily_bookings.plot(kind='line', marker='o', color='green')
plt.title('Daily Booking Trend')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# âœ… Step 19: Top 5 Pickup and Drop Locations

```python
# Top 5 Pickup Locations
top_pickups = df['Pickup_Location'].value_counts().head(5)
print("\nTop 5 Pickup Locations:\n", top_pickups)

# Top 5 Drop Locations
top_drops = df['Drop_Location'].value_counts().head(5)
print("\nTop 5 Drop Locations:\n", top_drops)

# Bar plot of top pickup and drop locations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
top_pickups.plot(kind='bar', color='teal')
plt.title('Top 5 Pickup Locations')
plt.ylabel('Number of Rides')

plt.subplot(1, 2, 2)
top_drops.plot(kind='bar', color='purple')
plt.title('Top 5 Drop Locations')
plt.ylabel('Number of Rides')

plt.tight_layout()
plt.show()
```

# âœ… Step 20: Compare Ratings (Driver vs Customer)

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df['Driver_Ratings'].dropna(), label='Driver Ratings', fill=True, color='blue')
sns.kdeplot(df['Customer_Rating'].dropna(), label='Customer Ratings', fill=True, color='orange')
plt.title('Distribution of Driver vs Customer Ratings')
plt.xlabel('Rating')
plt.legend()
plt.tight_layout()
plt.show()
```

# âœ… Step 21: Export Cleaned Dataset and Summary

```python
# Optional: Save the cleaned dataset to a new Excel file
cleaned_file_name = 'Cleaned_Ola_Ride_Data_July.xlsx'
df.to_excel(cleaned_file_name, index=False)
print(f"âœ… Cleaned dataset exported as: {cleaned_file_name}")

# Summary metrics for quick overview
print("\nðŸ”¹ Final Summary:")
print(f"Total Rides: {len(df)}")
print(f"Successful Rides: {(df['Booking_Status'] == 'Success').sum()}")
print(f"Canceled by Driver: {(df['Booking_Status'] == 'Canceled by Driver').sum()}")
print(f"Canceled by Customer: {(df['Booking_Status'] == 'Canceled by Customer').sum()}")
print(f"Average Booking Value: â‚¹{df['Booking_Value'].mean():.2f}")
print(f"Average Ride Distance: {df['Ride_Distance'].mean():.2f} km")
print(f"Average Driver Rating: {df['Driver_Ratings'].mean():.2f}")
print(f"Average Customer Rating: {df['Customer_Rating'].mean():.2f}")
```
